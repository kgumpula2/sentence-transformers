```python
import matplotlib.pyplot as plt
from matplotlib.axes import Axes
plt.style.use("ggplot")
```

```python
labels_dimred = [
  "baseline", 
  "truncation", 
  "sklearn-pca",
  "faiss-pca"
]
data_dimred = [
  [[1], [0.3051]],
  [[1/2, 1/4, 1/8], [0.2938, 0.2769, 0.2276]],
  [[1/2, 1/4, 1/8], [0.2997, 0.2741, 0.2034]],
  [[1/2, 1/4, 1/8], [0.2992, 0.2737, 0.2043]],
]
def apply_dimred_results(ax):
  ax: Axes
  ax.set_xscale("log")
  ax.minorticks_off()
  ax.invert_xaxis()
  ax.set_xticks([1, 1/2, 1/4, 1/8])
  ax.set_xticklabels(["1", "1/2", "1/4", "1/8"])
  ax.set_xlabel("Size Reduction")
  ax.set_ylabel("MRR @ 1000")
  ax.set_title("Dimensionality Reduction Score (mini-lm-384)")
  for i in range(len(data_dimred)):
    ax.plot(*data_dimred[i], label=labels_dimred[i], markevery = 1, marker = "o", linewidth=1, ms=4)
  ax.legend(prop={'size': 12})
```
